```python
import numpy as np
#Sigmoid Function
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_p(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
#How fast we want to adjust the learning rate
learningRate = .01

#Initialize the weights
w1 = np.random.randn()
w2 = np.random.randn()
w3 = np.random.randn()
b = np.random.randn()


print("Starting")

#Start the montage
for i in range(100000):
    #Get a random point in the data
    ri = np.random.randint(len(data))
    point = data[ri]
    
    #The equation for our line
    z = point[0] * w1 + point[1] * w2 + point[2] * w3 + b
    #Squish it into a sigmoid function
    pred = sigmoid(z)
    
    #Say that we want the target to be our testing value
    target = point[3]
    
    #Do a squared cost function to get a nice derivative of
    cost = np.square(pred - target)
    
    #Derivative of cost function with respect to pred
    dcost_pred = 2 * (pred - target)
    #Find the derivative of pred with respect to z
    dpred_dz = sigmoid_p(z)
    
    #Derivatives of z with respect to *
    dz_dw1 = point[0]
    dz_dw2 = point[1]
    dz_dw3 = point[2]
    dz_db = 1
    
    #Find the derivates of cost with respect to *
    dcost_dw1 = dcost_pred * dpred_dz * dz_dw1
    dcost_dw2 = dcost_pred * dpred_dz * dz_dw2
    dcost_dw3 = dcost_pred * dpred_dz * dz_dw3
    dcost_db = dcost_pred * dpred_dz * dz_db
    
    #Update the values
    w1 = w1 - learningRate * dcost_dw1
    w2 = w2 - learningRate * dcost_dw2
    w3 = w3 - learningRate * dcost_dw3
    b = b - learningRate * dcost_db
    
print("Done")
```

```python
def whichColor(red, green, blue):
    z = red * w1 + green * w2 + blue * w3 + b
    pred = sigmoid(z)
    if (pred >= .5):
        print("White")
    else:
        print("Black")
    print(pred)
```
